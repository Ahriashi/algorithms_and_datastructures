```python
f(n) = n*(1+1+1+1) + 2*n*(n-1)*(1+1+1+1) + 5 + 5
plot(f, (n,0,100), axes_labels=['n','f(n)'], 
     title='f(n) = n * (1+1+1+1) + 2 * n * (n-1) * (1+1+1+1) + 5 + 5', 
     title_pos=(0.5,-0.05))
```

```python
g(n) = n^2
t(n) = n
m(n) = 2^n

f_p = plot(f, (n,0,100), title='f(n)', title_pos=(0.5,-0.05))
t_p = plot(t, (n,0,100), title='t(n)', title_pos=(0.5,-0.05))
t_p
```
